```python
path = '/home/muody/data/synpuf/source/beneficiary/DE1_0_2008_Beneficiary_Summary_File_Sample_10.csv'
df = spark.read.csv(path, header=True)
df.printSchema()
df.head()

for col in df.columns:
    print("-"*80)
    print(col)
    print("Distinct values:", df.select(col).distinct().count())
    df.describe(col).show()
    df.groupby(col).agg({'MEDREIMB_IP': 'mean'}).show()
    print("-"*80)
```
